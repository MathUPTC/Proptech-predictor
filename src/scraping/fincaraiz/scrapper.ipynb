
Aca me he percatado que BeautifulSoup no cargaba todas las etiquetas "span" que tienen como clase "ant-typography", esto es debido a que hay una secci√≥n de la pagina (Comodidades adicionales) que se carga dinamicamente, por eso mismo se usar√° Selenium para poder acceder a dicha informaci√≥n

```python
# Configuraci√≥n Selenium

# Ruta al ejecutable de geckodriver (necesario para controlar Firefox)
ruta_driver = "/snap/bin/geckodriver"

# Se crea un servicio de Selenium usando el ejecutable especificado
service = Service(ruta_driver)

# Se inicia una nueva instancia del navegador Firefox con el servicio configurado
driver = webdriver.Firefox(service=service)

# Conexi√≥n con la p√°gina web deseada
driver.get("https://www.fincaraiz.com.co/apartamento-en-arriendo-en-tunja/192439336")

# Se espera 3 segundos para asegurar que el contenido de la p√°gina cargue
time.sleep(3)

try:
    # Se configura una espera expl√≠cita de hasta 10 segundos
    wait = WebDriverWait(driver, 10)

    # Se busca el bot√≥n "Ver m√°s" que est√° relacionado con "Comodidades de la propiedad"
    # Se utiliza un XPATH relativo que ubica un <h4> con cierto texto y luego busca el bot√≥n "Ver m√°s"
    boton = wait.until(EC.element_to_be_clickable((
        By.XPATH,
        "//h4[contains(text(), 'Comodidades de la propiedad')]/following::a[contains(@class, 'toggle-visibility')][1]/span[contains(text(), 'Ver m√°s')]"
    )))

    # Se hace scroll para que el bot√≥n sea visible en la ventana del navegador
    # El par√°metro 'block: center' centra el elemento en la vista
    driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", boton)

    # Se espera 1 segundo adicional para evitar problemas con headers flotantes u otras animaciones
    time.sleep(1)

    # Se hace clic en el bot√≥n usando JavaScript para evitar errores de superposici√≥n de elementos
    driver.execute_script("arguments[0].click();", boton)

    # Se espera 1 segundo para que el contenido expandido se cargue correctamente
    time.sleep(1)

# Si no se encuentra el bot√≥n o tarda demasiado en volverse clickeable, se captura la excepci√≥n
except (NoSuchElementException, TimeoutException):
    print("No se encontr√≥ el bot√≥n 'Ver m√°s'")

# Se captura todo el HTML actual de la p√°gina (incluyendo lo expandido) y se parsea con BeautifulSoup
vist_complet = BeautifulSoup(driver.page_source, "html.parser")
```

```python
#Funci√≥n Convertir Precio de str a int
def convertir_precio(texto):
    if not texto:
        return None
    try:
        # Eliminar s√≠mbolos y espacios
        texto_limpio = texto.replace('$', '').replace('.', '').replace(',', '').strip()
        return int(texto_limpio) if texto_limpio.isdigit() else None
    except (AttributeError, ValueError):
        return None

tarjetas = urlSoup.find_all("a", class_="lc-cardCover")

ruta_relativa = []
for link in tarjetas:
    ruta_relativa.append(link.get('href'))

ruta = [urljoin(url, href) for href in ruta_relativa if href is not None]

inmuebles = []

for i, enlace in enumerate(ruta):
    # Configuraci√≥n de headers para la petici√≥n HTTP
    headers = {
        'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:109.0) Gecko/20100101 Firefox/119.0',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8',
        'Accept-Language': 'es-ES,es;q=0.8,en-US;q=0.5,en;q=0.3',
        'Accept-Encoding': 'gzip, deflate',
        'Connection': 'keep-alive',
        'Upgrade-Insecure-Requests': '1'
    }

    # Extracci√≥n de datos iniciales con requests y BeautifulSoup
    try:
        response = requests.get(enlace, headers=headers)
        response.raise_for_status()
        ruta_Soup = BeautifulSoup(response.content, 'lxml')
        print("Conexi√≥n exitosa - P√°gina cargada correctamente\n")
        print(ruta_Soup.prettify())
    except requests.RequestException as e:
        print(f"Error en la petici√≥n: {e}")

    # Extracci√≥n de ubicaci√≥n
    contenedor_ubicacion_uno = ruta_Soup.find("div", class_="ant-col")
    todas_las_etiquetas = contenedor_ubicacion_uno.select("p.body.body-regular.medium")
    
    if len(todas_las_etiquetas) >= 2:
        s_etiqueta = todas_las_etiquetas[1]
        texto_final_uno = s_etiqueta.get_text(strip=True)
        print(texto_final_uno)
    else:
        print("No se encontr√≥ la etiqueta.")

    # Creaci√≥n del diccionario para el inmueble
    nuevo_inmueble = {
        "id": i+1,
        "precio": None,
        "ubicacion": None,
        "comodidades_adicionales": [],
        "descripcion": None
    }
    nuevo_inmueble["ubicacion"] = texto_final_uno

    # Extracci√≥n del precio
    elemento_precio = ruta_Soup.find("span", class_=lambda c: c and "price" in c.lower())
    precio = convertir_precio(elemento_precio.get_text(strip=True)) if elemento_precio else None
    nuevo_inmueble["precio"] = precio

    # Extracci√≥n de caracter√≠sticas t√©cnicas
    contenedor = ruta_Soup.find("div", class_="technical-sheet")
    hijos = contenedor.find_all("div", recursive=False)
    resultado = {}

    for hijo in hijos:
        spans = hijo.find_all("span", class_="ant-typography")
        llaves = [x.get_text(strip=True) for x in spans if x.get_text(strip=True) != "‚Ä¢"]

        strong_tag = hijo.find("strong")
        if strong_tag:
            valor = strong_tag.get_text(strip=True)
        else:
            boton = hijo.find("button")
            if boton:
                span_valor = boton.find("span")
                valor = span_valor.get_text(strip=True) if span_valor else None
            else:
                valor = None

        if llaves:
            clave = " | ".join(llaves)  
            resultado[clave] = valor

    nuevo_inmueble = nuevo_inmueble | resultado

    # Configuraci√≥n de Selenium para extraer contenido din√°mico
    ruta_driver = "/snap/bin/geckodriver"
    service = Service(ruta_driver)
    driver = webdriver.Firefox(service=service)
    driver.get(enlace)
    time.sleep(3)

    try:
        wait = WebDriverWait(driver, 10)
        boton = wait.until(EC.element_to_be_clickable((
            By.XPATH,
            "//h4[contains(text(), 'Comodidades de la propiedad')]/following::a[contains(@class, 'toggle-visibility')][1]/span[contains(text(), 'Ver m√°s')]"
        )))
        driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", boton)
        time.sleep(1)
        driver.execute_script("arguments[0].click();", boton)
        time.sleep(1)
    except (NoSuchElementException, TimeoutException):
        print("No se encontr√≥ el bot√≥n 'Ver m√°s'")

    vist_complet = BeautifulSoup(driver.page_source, "html.parser")
    driver.quit()

    # Extracci√≥n de comodidades
    titulo_comodidades = vist_complet.find('h4', string='Comodidades de la propiedad')
    comodidades = []

    if titulo_comodidades:
        contenedor_comodidades = titulo_comodidades.find_parent('div', class_='jsx-3623903256 container')
        
        if contenedor_comodidades:
            for span in contenedor_comodidades.find_all('span', class_='ant-typography'):
                if span.text.strip() != '‚Ä¢':
                    comodidades.append(span.text.strip())
            print("Comodidades encontradas:")
            print(comodidades)
        else:
            print("No se encontr√≥ el contenedor de comodidades")
    else:
        print("No se encontr√≥ la secci√≥n de comodidades")

    nuevo_inmueble["comodidades_adicionales"] = comodidades

    # Extracci√≥n de descripci√≥n
    descripcion_div = vist_complet.find('div', class_='ant-typography property-description body body-regular body-1 high')
    if descripcion_div:
        descripcion = descripcion_div.find('span').get_text(strip=True)
        print("Descripci√≥n del apartamento:")
        print(descripcion)
    else:
        print("No se encontr√≥ la secci√≥n de descripci√≥n")

    nuevo_inmueble["descripcion"] = descripcion
    inmuebles.append(nuevo_inmueble)
```

```python
url_dos = "https://www.fincaraiz.com.co/arriendo/casas-y-apartamentos/tunja/boyaca/pagina2"

headers = {
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:109.0) Gecko/20100101 Firefox/119.0',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8',
    'Accept-Language': 'es-ES,es;q=0.8,en-US;q=0.5,en;q=0.3',
    'Accept-Encoding': 'gzip, deflate',
    'Connection': 'keep-alive',
    'Upgrade-Insecure-Requests': '1'
}

try:
    response = requests.get(url_dos, headers=headers)
    response.raise_for_status()  # Lanza excepci√≥n si hay error HTTP

    url_dosSoup = BeautifulSoup(response.content, 'lxml')
    print("Conexi√≥n exitosa - P√°gina cargada correctamente\n")
    
    # Ver el HTML formateado
    print(url_dosSoup.prettify())
    
except requests.RequestException as e:
    print(f"Error en la petici√≥n: {e}")

tarjetas_dos = url_dosSoup.find_all("a", class_="lc-cardCover")

ruta_relativa_dos = []
for link in tarjetas_dos:
    ruta_relativa_dos.append(link.get('href'))

ruta_dos = [urljoin(url_dos, href) for href in ruta_relativa_dos if href is not None]






for i, enlace in enumerate(ruta_dos):
    # Configuraci√≥n de headers para la petici√≥n HTTP
    headers = {
        'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:109.0) Gecko/20100101 Firefox/119.0',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8',
        'Accept-Language': 'es-ES,es;q=0.8,en-US;q=0.5,en;q=0.3',
        'Accept-Encoding': 'gzip, deflate',
        'Connection': 'keep-alive',
        'Upgrade-Insecure-Requests': '1'
    }

    # Extracci√≥n de datos iniciales con requests y BeautifulSoup
    try:
        response = requests.get(enlace, headers=headers)
        response.raise_for_status()
        ruta_Soup = BeautifulSoup(response.content, 'lxml')
        print("Conexi√≥n exitosa - P√°gina cargada correctamente\n")
        print(ruta_Soup.prettify())
    except requests.RequestException as e:
        print(f"Error en la petici√≥n: {e}")

    # Extracci√≥n de ubicaci√≥n
    contenedor_ubicacion_uno = ruta_Soup.find("div", class_="ant-col")
    todas_las_etiquetas = contenedor_ubicacion_uno.select("p.body.body-regular.medium")
    
    if len(todas_las_etiquetas) >= 2:
        s_etiqueta = todas_las_etiquetas[1]
        texto_final_uno = s_etiqueta.get_text(strip=True)
        print(texto_final_uno)
    else:
        print("No se encontr√≥ la etiqueta.")

    # Creaci√≥n del diccionario para el inmueble
    nuevo_inmueble = {
        "id": i+1,
        "precio": None,
        "ubicacion": None,
        "comodidades_adicionales": [],
        "descripcion": None
    }
    nuevo_inmueble["ubicacion"] = texto_final_uno

    # Extracci√≥n del precio
    elemento_precio = ruta_Soup.find("span", class_=lambda c: c and "price" in c.lower())
    precio = convertir_precio(elemento_precio.get_text(strip=True)) if elemento_precio else None
    nuevo_inmueble["precio"] = precio

    # Extracci√≥n de caracter√≠sticas t√©cnicas
    contenedor = ruta_Soup.find("div", class_="technical-sheet")
    hijos = contenedor.find_all("div", recursive=False)
    resultado = {}

    for hijo in hijos:
        spans = hijo.find_all("span", class_="ant-typography")
        llaves = [x.get_text(strip=True) for x in spans if x.get_text(strip=True) != "‚Ä¢"]

        strong_tag = hijo.find("strong")
        if strong_tag:
            valor = strong_tag.get_text(strip=True)
        else:
            boton = hijo.find("button")
            if boton:
                span_valor = boton.find("span")
                valor = span_valor.get_text(strip=True) if span_valor else None
            else:
                valor = None

        if llaves:
            clave = " | ".join(llaves)  
            resultado[clave] = valor

    nuevo_inmueble = nuevo_inmueble | resultado

    # Configuraci√≥n de Selenium para extraer contenido din√°mico
    ruta_driver = "/snap/bin/geckodriver"
    service = Service(ruta_driver)
    driver = webdriver.Firefox(service=service)
    driver.get(enlace)
    time.sleep(3)

    try:
        wait = WebDriverWait(driver, 10)
        boton = wait.until(EC.element_to_be_clickable((
            By.XPATH,
            "//h4[contains(text(), 'Comodidades de la propiedad')]/following::a[contains(@class, 'toggle-visibility')][1]/span[contains(text(), 'Ver m√°s')]"
        )))
        driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", boton)
        time.sleep(1)
        driver.execute_script("arguments[0].click();", boton)
        time.sleep(1)
    except (NoSuchElementException, TimeoutException):
        print("No se encontr√≥ el bot√≥n 'Ver m√°s'")

    vist_complet = BeautifulSoup(driver.page_source, "html.parser")
    driver.quit()

    # Extracci√≥n de comodidades
    titulo_comodidades = vist_complet.find('h4', string='Comodidades de la propiedad')
    comodidades = []

    if titulo_comodidades:
        contenedor_comodidades = titulo_comodidades.find_parent('div', class_='jsx-3623903256 container')
        
        if contenedor_comodidades:
            for span in contenedor_comodidades.find_all('span', class_='ant-typography'):
                if span.text.strip() != '‚Ä¢':
                    comodidades.append(span.text.strip())
            print("Comodidades encontradas:")
            print(comodidades)
        else:
            print("No se encontr√≥ el contenedor de comodidades")
    else:
        print("No se encontr√≥ la secci√≥n de comodidades")

    nuevo_inmueble["comodidades_adicionales"] = comodidades

    # Extracci√≥n de descripci√≥n
    descripcion_div = vist_complet.find('div', class_='ant-typography property-description body body-regular body-1 high')
    if descripcion_div:
        descripcion = descripcion_div.find('span').get_text(strip=True)
        print("Descripci√≥n del apartamento:")
        print(descripcion)
    else:
        print("No se encontr√≥ la secci√≥n de descripci√≥n")

    nuevo_inmueble["descripcion"] = descripcion
    inmuebles.append(nuevo_inmueble)
```
